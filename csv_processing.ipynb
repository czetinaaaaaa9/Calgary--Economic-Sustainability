```python
import pandas as pd 
df = pd.read_csv("")
cols = df.columns.to_list() #prints all the csv's columns indexes in order to visualize data set and chose which columns must be kept 
```

```python
keep_idx = []
#after seeing the csv's columns insert what columns must be kept. Then eliminate them and see what the csv looks like now
df = df.iloc[:, keep_idx]
df.head()
```

```python
df = df.loc[df[''].notna()]
# eliminate 'na' entries for a specific column, can be repeated for any column.  
df
```

```python
df[''] = (df[''].replace(',', '', regex=True).astype(float))
# convert a columns values to float, eliminate commas if needed.
df[''] = df[''] * 0.0038706
#add  new column and define it by using a previous column. In our case the Property Assessement Column was used to compute the Property Tax column. 
df
```

```python

df = df.groupby(['']).sum()
# This groups all line groups all rows with the same column vaue (input) and sums the other columns. For example lots grouped by community code and their property values and taxes were added
df
```

```python
df_2 = pd.read_csv("",encoding='latin1')         ## open a new csv as a pandas dataframe. encoding used as there are many different types of csv files and this parameter is needed to deal with the different types.
df_2 = df_2.iloc[:,[,]]                          ## dataframe asigned only columns and rows specified 
df_2
```

```python
df_2 = df_2.loc[df_2[''] != 'x']
#this line can be used to keep only valid data entries. For example all the ones where the row's columnn is not 'x'
```

```python
# turn all rows from  column into float
df_2.iloc[:,1] = df_2.iloc[:,1].astype(float)
```

```python
# this block renames the indexes of one data set (df_2) using the naming convention of another one(df)
used_indices = set()

for code in df.index:
    # Match the code inside parentheses exactly
    pattern = f"\\({code}\\)"  # escape parentheses
    matching_indices = df_2[df_2[''].str.contains(pattern, regex=True, na=False)].index
    
    # Assign the code to the first unused matching row
    for idx in matching_indices:
        if idx not in used_indices:
            df_2.at[idx, "Community"] = code
            used_indices.add(idx)
            break  # stop after one-to-one match
print(df_2)
```

```python
# Export df_2
df_2.to_csv('', index=False)  # index=False to skip writing the row index

# Export df
df.to_csv('', index=True)  # keep COMM_CODE as index
```
